## Campgrounds 

```python
%%sql

select c.name, c.numberOfSitesFirstComeFirstServe from nps_public_data.campgrounds c
join nps_public_data.parks as p on p.id = c.parkId
limit 5;
```

```python
%%sql
SELECT
    p.fullname as park_name,
    c.name as campground_name

FROM nps_public_data.campgrounds c
INNER JOIN nps_public_data.parks p
    ON c.parkcode = p.parkcode
    AND p.designation = 'National Park'
WHERE p.name LIKE '%Yosemite%'
LIMIT 12;
```
